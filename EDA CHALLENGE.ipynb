Extraer los datos de los diferentes csv

```python
import pandas as pd
import cufflinks as cf
import plotly as plt
import seaborn as sns 
from IPython.display import display,HTML
from zipfile import ZipFile
```

```python
clarity = pd.DataFrame(pd.read_csv('./data/diamonds_clarity.csv'))
clarity
```

```python
city = pd.DataFrame(pd.read_csv('./data/diamonds_city.csv'))
city
```

```python
color = pd.DataFrame(pd.read_csv('./data/diamonds_color.csv'))
color
```

```python
cut = pd.DataFrame(pd.read_csv('./data/diamonds_cut.csv'))
cut
```

```python
dimensions = pd.DataFrame(pd.read_csv('./data/diamonds_dimensions.csv'))
dimensions
```

```python
properties = pd.DataFrame(pd.read_csv('./data/diamonds_properties.csv'))
properties.head
```

```python
dimensions_porporties = pd.merge(dimensions, properties)
dimensions_porporties
```

```python
transactional = pd.DataFrame(pd.read_csv('./data/diamonds_transactional.csv'))
transactional
```

```python
df_cut = pd. merge(dimensions_porporties, cut)
df_cut 
```

```python
df_color = pd.merge(df_cut, color)
df_color
```

```python
df_clarity = pd.merge(df_color, clarity)
df_clarity
```

```python
df_transactional = pd.merge(transactional, df_clarity)
df_transactional
```

```python
df_city = pd.merge(df_transactional, city)
df_city
```

```python
df_diamonds = df_city[['index_id', 'price', 'carat', 'depth','table','x','y','z','city','color','clarity','cut']]
df_diamonds
```

```python
by_price = df_diamonds.sort_values('price',ascending=False)
by_price 
```

# Price

```python
#Frecuency per Price 

import numpy as np
n = df_diamonds.shape[0]
k = 1 +(10/3) * np.log10(n)
k = int(k.round(0))
k
```

```python
frecuencia = pd.value_counts(pd.cut(x = df_diamonds.price, bins = k, include_lowest = True), sort = False)
frecuencia
```

# City

```python
import plotly.express as px
cf.set_config_file(sharing='public',theme='polar', offline=True)
```

```python
city_count = df_diamonds.city.value_counts()
city_perc =df_diamonds['city'].value_counts(normalize = True)*100
dis_frec_city = pd.DataFrame({'Count': city_count, '%':city_perc})
dis_frec_city
```

```python
city_count = df_diamonds['city'].value_counts()
colores = ['skyblue', 'plum', 'lightpink', 'lightgrey', 'wheat', 'lightblue', 'thistle']

# Crear el gráfico de barras con colores personalizados y ajustar el tamaño
plt.figure(figsize=(8, 5))
bars = city_count.plot(kind='bar', color=colores)
plt.title('Frequency of Cities', fontsize=14)
plt.xlabel('City', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor visibilidad
# Añadir etiquetas en cada barra
for i, count in enumerate(city_count):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom', fontsize=8)
mean_value = np.mean(city_count)
plt.axhline(y=mean_value, color='blue', linestyle='--', label='Mean')
plt.legend()
# Mostrar el gráfico
plt.show()
```
